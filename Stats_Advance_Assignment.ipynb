**Theory**

1. **What is a random variable in probability theory?**
   - A **random variable** is a variable whose values are determined by the outcome of a random event or experiment. It can take different values based on the probabilities of different outcomes.

2. **What are the types of random variables?**
   - There are two main types of random variables:
     - **Discrete random variable**: Takes on a finite or countably infinite set of distinct values (e.g., number of heads in coin flips).
     - **Continuous random variable**: Takes on any value within a given range (e.g., height, weight, time).

3. **What is the difference between discrete and continuous distributions?**
   - **Discrete distributions** deal with random variables that can take only distinct, countable values (like integers), such as the **Poisson** or **Binomial** distributions.
   - **Continuous distributions** deal with random variables that can take any value in a continuous range, like the **Normal** or **Uniform** distributions.

4. **What are probability distribution functions (PDF)?**
   - A **probability distribution function (PDF)** describes the likelihood of a random variable taking a particular value. For **discrete variables**, it's represented by a probability mass function (PMF), and for **continuous variables**, it’s the function whose integral over a range gives the probability.

5. **How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?**
   - A **cumulative distribution function (CDF)** gives the probability that a random variable will take a value less than or equal to a specific point. The **PDF** gives the probability of the variable taking an exact value (for discrete variables) or the likelihood per unit of the continuous variable.
   - CDF is the integral (for continuous variables) or sum (for discrete variables) of the PDF.

6. **What is a discrete uniform distribution?**
   - A **discrete uniform distribution** is a probability distribution where all outcomes are equally likely. For example, when rolling a fair die, each of the six outcomes (1 through 6) has an equal probability of occurring.

7. **What are the key properties of a Bernoulli distribution?**
   - The **Bernoulli distribution** describes experiments with exactly two outcomes: success (usually denoted as 1) and failure (denoted as 0). The key properties are:
     - There is only one trial (n = 1).
     - The probability of success is denoted by \( p \), and the probability of failure is \( 1 - p \).

8. **What is the binomial distribution, and how is it used in probability?**
   - The **binomial distribution** models the number of successes in a fixed number of independent Bernoulli trials. It is used when there are two possible outcomes (success/failure) in a set number of trials. The probability of getting exactly \( k \) successes in \( n \) trials is given by:
     \[
     P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}
     \]
   - Where \( p \) is the probability of success, and \( \binom{n}{k} \) is the binomial coefficient.

9. **What is the Poisson distribution and where is it applied?**
   - The **Poisson distribution** models the number of events that occur within a fixed interval of time or space, given that these events happen independently and at a constant rate. It’s widely used for rare events like the number of car accidents in a day or phone calls to a call center.

10. **What is a continuous uniform distribution?**
    - A **continuous uniform distribution** represents a random variable that has an equal probability of taking any value within a specified range. For example, the probability of getting a specific time between 0 and 1 second is uniform in that interval.

11. **What are the characteristics of a normal distribution?**
    - The **normal distribution** is symmetric and bell-shaped. Its key characteristics include:
      - Mean \( \mu \) and standard deviation \( \sigma \).
      - The area under the curve is 1 (it represents total probability).
      - Approximately 68% of values lie within one standard deviation, 95% within two, and 99.7% within three (68-95-99.7 rule).

12. **What is the standard normal distribution, and why is it important?**
    - The **standard normal distribution** is a normal distribution with a mean of 0 and a standard deviation of 1. It's important because many statistical methods rely on standardizing data to this form, allowing comparisons across different datasets and easier use of Z-scores.

13. **What is the Central Limit Theorem (CLT), and why is it critical in statistics?**
    - The **Central Limit Theorem** states that, regardless of the distribution of the original population, the sampling distribution of the sample mean approaches a normal distribution as the sample size increases. This is critical because it allows us to make inferences about population parameters using normal distribution approximations, even if the population itself is not normally distributed.

14. **How does the Central Limit Theorem relate to the normal distribution?**
    - The CLT ensures that the distribution of sample means tends to be normal, even if the underlying population distribution is not normal. This is why normal distribution plays a key role in many statistical procedures and hypothesis testing.

15. **What is the application of Z statistics in hypothesis testing?**
    - **Z statistics** (or Z-scores) are used in hypothesis testing to determine how far a sample statistic is from the population parameter in terms of standard deviations. A Z-test is commonly used to test if a sample mean significantly differs from the population mean, assuming the population is normally distributed.

16. **How do you calculate a Z-score, and what does it represent?**
    - A **Z-score** is calculated by subtracting the population mean from the sample value and dividing the result by the population standard deviation:
      \[
      Z = \frac{X - \mu}{\sigma}
      \]
    - It represents how many standard deviations a value is from the mean.

17. **What are point estimates and interval estimates in statistics?**
    - **Point estimates** provide a single value as an estimate of a population parameter (e.g., sample mean as an estimate of population mean).
    - **Interval estimates** give a range of values within which the parameter is likely to lie, typically expressed as confidence intervals.

18. **What is the significance of confidence intervals in statistical analysis?**
    - **Confidence intervals** provide a range of values, derived from the sample data, that are likely to contain the population parameter with a specified level of confidence (e.g., 95%). They are important for understanding the precision and reliability of an estimate.

19. **What is the relationship between a Z-score and a confidence interval?**
    - A **Z-score** is used to calculate the margin of error in constructing a **confidence interval**. For example, for a 95% confidence level, the Z-score corresponds to a critical value that helps determine the interval around the sample mean.

20. **How are Z-scores used to compare different distributions?**
    - Z-scores standardize values from different distributions, allowing you to compare them even if the distributions have different means and standard deviations. This way, values from different distributions can be assessed on the same scale.

21. **What are the assumptions for applying the Central Limit Theorem?**
    - The CLT assumes:
      - The samples are independent.
      - The sample size is large enough (typically \( n \geq 30 \)).
      - The underlying population has finite variance and mean.

22. **What is the concept of expected value in a probability distribution?**
    - The **expected value (or mean)** of a random variable is the long-term average value of repetitions of the experiment. It is calculated as the sum of all possible values weighted by their probabilities for discrete variables or the integral for continuous variables.

23. **How does a probability distribution relate to the expected outcome of a random variable?**
    - A **probability distribution** gives the probabilities of all possible outcomes of a random variable, while the **expected value** is the weighted average of all those possible outcomes, giving the central or "typical" value you can expect in the long run.

```python
#6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.
import numpy as np
import matplotlib.pyplot as plt

def discrete_uniform_cdf(low, high):
    outcomes = np.arange(low, high + 1)
    cdf_values = (outcomes - low + 1) / (high - low + 1)  # CDF calculation
    return outcomes, cdf_values

# Set the parameters for the discrete uniform distribution
low = 1
high = 6

# Calculate the CDF
x_values, cdf_values = discrete_uniform_cdf(low, high)

# Plot the CDF
plt.step(x_values, cdf_values, where='post', color='skyblue', linewidth=2, label="CDF")
plt.title('Cumulative Distribution Function (CDF) of Discrete Uniform Distribution')
plt.xlabel('Outcome')
plt.ylabel('Cumulative Probability')
plt.xticks(np.arange(low, high + 1))
plt.grid(True)
plt.legend()
plt.show()
```

```python
#19. Implement hypothesis testing using Z-statistics for a sample dataset.
import numpy as np
from scipy.stats import norm

def z_test(sample_mean, population_mean, sample_std_dev, sample_size, significance_level):
    z_score = (sample_mean - population_mean) / (sample_std_dev / np.sqrt(sample_size))
    p_value = 2 * (1 - norm.cdf(np.abs(z_score)))
    if p_value < significance_level:
        print("Reject the null hypothesis.")
    else:
        print("Fail to reject the null hypothesis.")

sample_mean = 10.5
population_mean = 10
sample_std_dev = 2
sample_size = 50
significance_level = 0.05
z_test(sample_mean, population_mean, sample_std_dev, sample_size, significance_level)

```

```python
#20. Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence_level):
    sample_mean = np.mean(data)
    sample_std_dev = np.std(data, ddof=1)
    sample_size = len(data)
    critical_value = stats.t.ppf(1 - (1 - confidence_level) / 2, df=sample_size - 1)
    margin_of_error = critical_value * (sample_std_dev / np.sqrt(sample_size))
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    return lower_bound, upper_bound

data = np.array([10, 15, 20, 25, 30])
confidence_level = 0.95
lower, upper = confidence_interval(data, confidence_level)
print(f"95% Confidence Interval: ({lower}, {upper})")
```

```python
#21. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence_level=0.95):
    n = len(data)
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)

    t_critical = stats.t.ppf((1 + confidence_level) / 2, df=n - 1)

    # Margin of error
    margin_of_error = t_critical * (sample_std / np.sqrt(n))

    # Confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Parameters for normal distribution
mean = 50  # True mean
std_dev = 10  # Standard deviation
sample_size = 1000  # Number of samples

# Generate data from a normal distribution
data = np.random.normal(mean, std_dev, sample_size)

# Calculate the confidence interval for the mean
confidence_level = 0.95
lower, upper = confidence_interval(data, confidence_level)

# Display the results
print(f"Sample Mean: {np.mean(data)}")
print(f"{confidence_level * 100}% Confidence Interval for the Mean: ({lower}, {upper})")
```

```python
#22. Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def normal_distribution_pdf(x, mean, std_dev):
    pdf = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev)**2)
    return pdf

mean = 0
std_dev = 1
x = np.linspace(-4, 4, 1000)
y = normal_distribution_pdf(x, mean, std_dev)
plt.plot(x, y, color='skyblue')
plt.title('Normal Distribution PDF')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```

```python
#23. Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

def plot_poisson_cdf(lmbda, max_x):

    x_values = np.arange(0, max_x + 1)
    cdf_values = poisson.cdf(x_values, lmbda)


    plt.step(x_values, cdf_values, where='post', color='skyblue', linewidth=2, label=f'Poisson CDF (λ={lmbda})')
    plt.title(f'Cumulative Distribution Function (CDF) of Poisson Distribution')
    plt.xlabel('Number of Events (x)')
    plt.ylabel('Cumulative Probability')
    plt.grid(True)
    plt.legend()
    plt.show()

lambda_value = 4
max_x = 15

# Plot the Poisson CDF
plot_poisson_cdf(lambda_value, max_x)
```

```python
#25. Write a Python program to compare the standard deviations of two datasets and visualize the difference.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

dataset1 = np.array([10, 15, 20, 25, 30])
dataset2 = np.array([5, 10, 15, 20, 30])

std_dev_1 = np.std(dataset1)
std_dev_2 = np.std(dataset2)


print(f"Standard Deviation of Dataset 1: {std_dev_1:.2f}")
print(f"Standard Deviation of Dataset 2: {std_dev_2:.2f}")

# Plot the datasets for comparison
plt.figure(figsize=(12, 6))

# Plot the first dataset
plt.subplot(1, 2, 1)
plt.hist(dataset1, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title(f'Dataset 1 (Std = {std_dev_1:.2f})')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Plot the second dataset
plt.subplot(1, 2, 2)
plt.hist(dataset2, bins=30, color='orange', edgecolor='black', alpha=0.7)
plt.title(f'Dataset 2 (Std = {std_dev_2:.2f})')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plots
plt.tight_layout()
plt.show()

# Bar chart to visualize the difference in standard deviations
plt.bar(['Dataset 1', 'Dataset 2'], [std_dev_1, std_dev_2], color=['skyblue', 'orange'])
plt.title('Comparison of Standard Deviations')
plt.ylabel('Standard Deviation')
plt.show()

```

```python
#27. Implement Z-score normalization on a dataset and visualize its transformation.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def z_score_normalization(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev
    return z_scores

data = np.array([10, 15, 20, 25, 30])
normalized_data = z_score_normalization(data)

plt.hist(normalized_data, bins=30, color='orange', edgecolor='black', alpha=0.7)
plt.title('Z-Score Normalized Data')
plt.xlabel('Z-Score')
plt.ylabel('Frequency')
plt.show()
```

```python
#28. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_and_kurtosis(data):
    skewness = skew(data)
    kurtosis_value = kurtosis(data)
    return skewness, kurtosis_value

data = np.random.normal(loc=0, scale=1, size=1000)
skewness, kurtosis_value = calculate_skewness_and_kurtosis(data)
print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurtosis_value}")
```
